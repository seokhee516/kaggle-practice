# 'data-preparation-exploration' 커널과 거의 유사

- 변수명 직관적으로 변경
- 다양한 시각화 도구 사용

```python
data = []
for feature in trainset.columns:
    # 데이터 역할 지정
    if feature == 'target':
        use = 'target'
    elif feature == 'id':
        use = 'id'
    else:
        use = 'input'
         
    # 데이터 type 지정
    if 'bin' in feature or feature == 'target':
        type = 'binary'
    elif 'cat' in feature or feature == 'id':
        type = 'categorical'
    elif trainset[feature].dtype == float or isinstance(trainset[feature].dtype, float):
        # float 타입
        type = 'real'
    elif trainset[feature].dtype == int:
        type = 'integer'
    # id는 False로 지정해주어 버리기로 하고, 나머지 preserve는 True로 가져감
    preserve = True
    if feature == 'id':
        preserve = False
    
    # data type 정의 
    dtype = trainset[feature].dtype
    
    category = 'none'
    # category 정의
    if 'ind' in feature:
        category = 'individual'
    elif 'reg' in feature:
        category = 'registration'
    elif 'car' in feature:
        category = 'car'
    elif 'calc' in feature:
        category = 'calculated'
    
    
    # metadata 만들기 위해 딕셔너리 만듬
    feature_dictionary = {
        'varname': feature,
        'use': use,
        'type': type,
        'preserve': preserve,
        'dtype': dtype,
        'category' : category
    }
    data.append(feature_dictionary)
    
metadata = pd.DataFrame(data, columns=['varname', 'use', 'type', 'preserve', 'dtype', 'category'])
metadata.set_index('varname', inplace=True)
metadata
```

# data quality 확인
결측값 확인


```python
vars_with_missing = []

for feature in trainset.columns:
  missings = trainset[trainset[feature]==-1][feature].count()
  if missings > 0:
    # 결측치 있다면 vars_with_missing 리스트에 추가
    vars_with_missing.append(feature)
    # 결측 비율 계산
    missing_perc = missings/trainset.shape[0]
    print('{} 변수는 {}개의 결측치({:.2%})를 가진다.'.format(feature, missings, missing_perc))
print('총 {}개의 변수가 결측치를 가진다.'.format(len(vars_with_missing)))
```
